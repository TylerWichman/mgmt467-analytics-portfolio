### Q2 — Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*  

For each chart, list: **Question archetype**, **Data shape**, **Pitfalls**, **Alternatives**.

📝 **Your response (template):**
1. **Bar:**  
   - Question archetype: How do categories compare?
   - Data shape: categorical (x-axis) vs. numerical (y-axis)
   - Pitfalls:  Using too many categories, starting at a non zero baseline on axis
   - Alternatives:  Stacked bar charts, heatmaps
2. **Line:**  
   - Question archetype:  How does something change based on a variable?
   - Data shape:  continuous x-axis vs numerical y-axis
   - Pitfalls:  misleading scales, non-continuous data
   - Alternatives: Area charts, heatmap calander
3. **Scatter:**
   - Question archetype: How do two variables relate?
   - Data shape: numerical vs numerical
   - Pitfalls: Assuming correlation = causation, ignoring confounding variables
   - Alternatives: Bubble chart, Pair plot
4. **Histogram:**  
   - Question archetype: What is the distribution of a variable?
   - Data shape:  Numerical variable grouped into bins
   - Pitfalls: bin sizing, doesnt show outliers well
   - Alternatives: Box Plot, KDE
5. **Map:**  
   - Question archetype: How does geography influence a variable?
   - Data shape: location mapped to a numerical variable
   - Pitfalls: areas differ in size, missing data exagerrates patterns
   - Alternatives: Cartograms, heat maps

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

df_flights.head()
````

````python
on_time_rates = (
    df_flights.groupby("carrier")["on_time"]
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

# Convert to percentage
on_time_rates["on_time_pct"] = on_time_rates["on_time"] * 100

# Plot
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))

ax = sns.barplot(
    data=on_time_rates,
    x="carrier",
    y="on_time_pct",
    palette="Blues_d"
)

# Add percentage labels on bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%",
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10)

# Titles and labels
ax.set_title("On-Time Rate by Carrier", fontsize=14, weight="bold")
ax.set_ylabel("On-Time Rate (%)", fontsize=12)
ax.set_xlabel("Carrier", fontsize=12)

# Tidy up layout
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
daily_delay = (
    df_flights.groupby("date")["arr_delay_min"]
    .mean()
    .reset_index()
    .sort_values("date")
)

# Compute 7-day rolling average
daily_delay["rolling_7d"] = daily_delay["arr_delay_min"].rolling(7).mean()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(daily_delay["date"], daily_delay["arr_delay_min"],
         label="Daily Avg Delay", color="skyblue", linewidth=1.5)
plt.plot(daily_delay["date"], daily_delay["rolling_7d"],
         label="7-Day Rolling Avg", color="red", linewidth=2)

# Labels, title, grid
plt.title("Average Daily Arrival Delay with 7-Day Rolling Average", fontsize=14, weight="bold")
plt.xlabel("Date", fontsize=12)
plt.ylabel("Average Delay (minutes)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram data
delays = df_flights["arr_delay_min"].dropna()

# % of flights within [-5, 15]
mask = delays.between(-5, 15)
pct_in_window = 100 * mask.mean()

# Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(delays, bins=50, kde=False, color="steelblue")

# Reference line at 0
plt.axvline(0, color="red", linestyle="--", linewidth=1.5, label="On Time (0 min)")

# Annotate percentage
plt.text(0.98, 0.95,
         f"{pct_in_window:.1f}% within [-5, 15] min",
         transform=plt.gca().transAxes,
         ha="right", va="top", fontsize=12,
         bbox=dict(facecolor="white", alpha=0.7, edgecolor="gray"))

# Labels and title
plt.title("Distribution of Arrival Delays", fontsize=14, weight="bold")
plt.xlabel("Arrival Delay (minutes)", fontsize=12)
plt.ylabel("Number of Flights", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.tight_layout()
plt.show()
````

````python
df_routes = (
    df_flights.groupby(["origin", "dest"])["load_factor"]
    .mean()
    .reset_index()
)

# Create a route label like "JFK → LAX"
df_routes["route"] = df_routes["origin"] + " → " + df_routes["dest"]

# Top 15 routes by mean load factor
top_routes = (
    df_routes.sort_values("load_factor", ascending=False)
    .head(15)
    .sort_values("load_factor")  # resort for horizontal bar order
)

# Plot horizontal bar chart
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

ax = sns.barplot(
    data=top_routes,
    x="load_factor",
    y="route",
    palette="Blues_d"
)

# Labels and title
ax.set_title("Top 15 Routes by Average Load Factor", fontsize=14, weight="bold")
ax.set_xlabel("Average Load Factor", fontsize=12)
ax.set_ylabel("Route", fontsize=12)

# Format x-axis as percentage if values are 0–1
if top_routes["load_factor"].max() <= 1.2:
    ax.set_xlim(0, 1)
    ax.set_xticklabels([f"{x:.0%}" for x in ax.get_xticks()])

plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    df_flights["cost_usd"],
    df_flights["revenue_usd"],
    c=df_flights["casm"],
    cmap="viridis",
    alpha=0.7,
    edgecolor="k",
    linewidth=0.5
)

# Diagonal y=x line
lims = [
    min(df_flights["cost_usd"].min(), df_flights["revenue_usd"].min()),
    max(df_flights["cost_usd"].max(), df_flights["revenue_usd"].max())
]
plt.plot(lims, lims, "r--", linewidth=1.5, label="y = x")

# Labels and title
plt.title("Cost vs. Revenue (colored by CASM)", fontsize=14, weight="bold")
plt.xlabel("Cost (USD)", fontsize=12)
plt.ylabel("Revenue (USD)", fontsize=12)

# Colorbar for CASM
cbar = plt.colorbar(scatter)
cbar.set_label("CASM", fontsize=12)

# Legend for reference line
plt.legend()

# Grid and layout
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
````
